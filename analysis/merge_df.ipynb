```python
import pandas as pd
import os
```

```python
# Define the paths to the datasets
data_dir = '../Data/Feature_Engineered'
clientes_path = os.path.join(data_dir, 'hey_clientes_engineered.csv')
portfolio_path = os.path.join(data_dir, 'portfolio_features.csv')
transactions_path = os.path.join(data_dir, 'user_features_transactions.csv')

# Load the datasets
clientes_df = pd.read_csv(clientes_path)
portfolio_df = pd.read_csv(portfolio_path)
transactions_df = pd.read_csv(transactions_path)

print('Clientes dataset shape:', clientes_df.shape)
print('Portfolio dataset shape:', portfolio_df.shape)
print('Transactions dataset shape:', transactions_df.shape)
```

```python
# Check the first few rows of each dataset to confirm user_id is present
print('Clientes columns:', clientes_df.columns.tolist())
print('Portfolio columns:', portfolio_df.columns.tolist())
print('Transactions columns:', transactions_df.columns.tolist())
```

```python
# Merge clientes and portfolio on user_id
merged_df = pd.merge(clientes_df, portfolio_df, on='user_id', how='inner')
print('After merging clientes and portfolio:', merged_df.shape)
```

```python
# Merge the result with transactions on user_id
final_df = pd.merge(merged_df, transactions_df, on='user_id', how='inner')
print('After merging with transactions:', final_df.shape)
```

```python
# Display the first few rows of the merged dataset
final_df.head()
```

```python
# Save the merged dataset
output_path = os.path.join(data_dir, 'merged_features.csv')
final_df.to_csv(output_path, index=False)
print(f'Merged dataset saved to {output_path}')
```
